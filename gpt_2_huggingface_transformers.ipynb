O _GPT2Tokenizer_ vai pegar o texto de entrada e codificá-lo na forma de números. Os números serão passados para o modelo GPT-2. No final o _GPT2Tokenizer_ fará a operação inversa, irá transformar os números de saída do modelo em texto novamente.

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

#1. Carregar o modelo

```python
# carrega o tokenizer com o modelo pré-treinado
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
# carrega o modelo pré-treinado. pad_token_id representa qual token será usado para preencher o texto e o tokenizer.eos_token_id representa o token de final de frase.
model = GPT2LMHeadModel.from_pretrained('gpt2-large', pad_token_id=tokenizer.eos_token_id)
```

#2. Tokenizar uma sentença

```python
# cria uma string (sentença) para entrada no modelo
sentence = "what is thermodynamics?"
#tokenizar -> codificar a frase de números que identificam as palavras da sentença. No final o retorno de tensores é do tipo tensor pytorch
input_ids = tokenizer.encode(sentence, return_tensors='pt')
```

#3. Gerar o texto e decodificá-lo

```python
# max_length -> máxima quantidade de palavras no texto gerado.
# num_beams -> método beans search para encontrar a próxima palavra mais adequada na sequência. O número de feixes (beans) é quantas árvores de pesquisa será utilizada.
# no_repeat_ngram_size -> impede que o modelo repita certas sequeências. Com o número 2 a sequência só pode ser repetida uma vez.
# early_stopping -> não estiveram melhorando os resultados, o modelo vai parar de gerar
output = model.generate(input_ids, max_length=4096, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

#4. Resultado

```python
# decode para decodificar a saída. Números -> Texto
# output[0] -> pega a primeira parte da matriz
# skip_special_tokens = True -> para não exibir tokens no final de frase, nem tokens especiais
text = tokenizer.decode(output[0], skip_special_tokens=True)
print (text)
```
